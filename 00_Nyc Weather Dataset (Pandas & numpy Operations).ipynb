```python
import pandas as pd
India_weather = pd.DataFrame({
    "city" : ["delhi", "mumbai", "chennai"],
    "temp" : [3,4,5],
    "humidity" : [30,45,66]
})
India_weather
```

```python
df = pd.concat([India_weather,us_weather])   # with same index
df
```

```python
df = pd.concat([India_weather,us_weather], ignore_index = True)  #with different index
df
```

```python
df = pd.concat([India_weather,us_weather], axis = 1)
df
```

```python
def binary_search(arr, low, high, x):
# Check base case
	if high >= low:
		mid = (high + low) // 2
		if arr[mid] == x:
			return mid
		elif arr[mid] > x:
			return binary_search(arr, low, mid - 1, x)
		# Else the element can only be present in right subarray
		else:
			return binary_search(arr, mid + 1, high, x)
	else:
		# Element is not present in the array
         return -1
# Test array
arr = [ 2, 3, 4, 10, 40 ]
x = 10
# Function call
result = binary_search(arr, 0, len(arr)-1, x)
if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in array")
```

```python
import random          #using normal comapare time complexity O(m*n)
l1 =[1,2,3,4,5,6,7,8]
random.shuffle(l1)
l2 =[5,6,10,35,64,33,89,87]
random.shuffle(l2)
count = 0;
for i in l1:
    for j in l2:
        if i == j:
            print(i)
            count += 1;
print("number of common elements is ", count)                
```

```python
import random        #using hashnap or hashtable dictionary   time com. O(m+n)
l1 =[1,2,3,4,5,6,7,8]
random.shuffle(l1)
l2 =[5,6,10,35,64,33,89,87]
random.shuffle(l2)
SmallList ={}
for ele in l2:
    SmallList[ele] = 1;
count = 0;
for i in l1:
    if SmallList.get(i) != None:
            print(i)
            count += 1;
print("number of common elements is ", count)  
```
